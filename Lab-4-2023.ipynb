## Exercise 1
* Read the dataset Titanic, create the dataframe and read all columns.
* Create a bar chart and pie chart for the column Age, Parch and Sex of the Titanic dataset.

```python
df = pd.read_csv("Titanic.csv")
```

```python
df.head()
```

```python
df["Age"].iloc[0:10].plot(kind='bar')
```

```python
df.groupby('Parch').Age.mean().plot(kind='bar')
```

```python
x = df.Sex
y = df.Age
plt.bar(x,y)
```

```python
df['Sex'].value_counts().plot.pie(autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
df['Parch'].value_counts().plot.pie(autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels = labels, autopct = '%1.1f%%',shadow = True, startangle = 90)
ax1.axis('equal')

plt.show()
```

## Exercise 4


Add some text to a graph & create and plot a random linear graph
Right at the center if the graph add some text which says this is the center.
Also add grids to the graph.

```python
A = np.linspace(0, 10, 100)
B = A + np.random.normal(0, 1, 100)
plt.plot(A, B)
plt.text(5, 5, 'This is the center', ha = 'center', va = 'center', fontsize = 16, color='green')
plt.grid(True)
plt.show()
```
