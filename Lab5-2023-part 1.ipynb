## Exercises 6


Create a Categorical Plot for the column Sex of the Titanic dataset.

```python
data = pd.read_csv("Titanic.csv")
data.head()
```

```python
sns.countplot(x = 'Sex', data = data, palette = "flare")
plt.title('Count of Passengers by Sex')
plt.show()
```

```python
sns.countplot(x = 'Sex', data = data, hue = "Pclass", palette = "light:#5A9")
plt.title('Count of Passengers by Sex with respect to Pclass')
plt.show()
```

```python
sns.countplot(x = 'Sex', data = data, hue = "Embarked", palette = "ch:s=.25,rot=-.25")
plt.title('Count of Passengers by Sex with respect to embarked')
plt.show()
```
