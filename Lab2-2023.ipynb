What is the average age of the 5 oldest passengers? The reset_index method will be helpful here.

```python
#Write your code here
df1 = df["Age"].sort_values(ascending = False)
age = []
for i in df1[0:5]:
    age.append(i)
print("5 Oldest Peoples Age:",age)
SumofAges = 0
for n in age:
     SumofAges = SumofAges + n
AverageAge = SumofAges/5
print("Average age of the 5 oldest passengers:",AverageAge)
```
