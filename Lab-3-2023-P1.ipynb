## Question 1

Follow steps: 
* Create a figure object called fig using plt.figure() **
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. 
* Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1]) 

ax.plot(x,y,"r")
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('X vs Y')
```

## Question 2
* Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.
* plot (x,y) on both axes. And call your figure object to show it.

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 2, 2]) 

ax1.plot(x,y,"r")
ax1.set_xlabel('X Label')
ax1.set_ylabel('Y Label')
ax1.set_title('X vs Y')

ax2.plot(y, x, 'r')
ax2.set_xlabel('X_Label')
ax2.set_ylabel('Y_Label')
ax2.set_title('Y vs X');
```

## Question 5


Consider the x=np.arange(0,15,0.5),
them plot (np.sin(x)), (np.sin(x+0.5)), (np.sin(x+1.0)), (np.cos(x))
with different linestyle and linewidth.


```python
x = np.arange(0,15,0.5)
fig, ax = plt.subplots()
ax.plot(np.sin(x), 'r.-') # blue line with dots
ax.plot(np.sin(x+0.5), 'cyan', lw = '2', ls = '--', marker = '1') # green dashed line
ax.plot(np.sin(x+1.0), color = 'orange', lw = 3, ls = '-', marker = 'o')
ax.plot(np.cos(x), color = 'purple', lw = 5, ls = '--', marker = 's')
```
